```python
import numpy as np
import cv2
from utils import color_hist
```

Canny edge detection

```python
# cap = cv2.VideoCapture('../nocutcorners.mp4')
cap = cv2.VideoCapture(0)

while(1):
    _, frame = cap.read()
    if not _:
        break
    
    
    # doesn't improve when using gray colored img
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    gray = cv2.equalizeHist(gray)
    edges_g = cv2.Canny(gray, 150, 200)
    edges_c = cv2.Canny(frame, 100, 200)
    cv2.imshow("edges_g", edges_g)
    cv2.imshow("edges_c", edges_c)
    
    k = cv2.waitKey(60) & 0xff
    if k == 27:
        break

cv2.destroyAllWindows()
cap.release()
```

Hough lines

```python
cap = cv2.VideoCapture('../nocutcorners.mp4')
# cap = cv2.VideoCapture(0)

while(1):
    _, frame = cap.read()
    if not _:
        break
        
    img = frame.copy()
    # doesn't improve when using gray colored img
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    edges = cv2.Canny(gray, 150, 200, apertureSize = 3)
    
    minLineLength = 100
    maxLineGap = 100
    lines = cv2.HoughLinesP(edges,1,np.pi/180,100,minLineLength,maxLineGap)
    if lines is not None:
        for x1,y1,x2,y2 in lines[0]:
            cv2.line(img,(x1,y1),(x2,y2),(0,255,0),2)

    cv2.imshow("edges", edges)
    cv2.imshow("lines", img)
    
    k = cv2.waitKey(60) & 0xff
    if k == 27:
        break

cv2.destroyAllWindows()
cap.release()
```
