```python
import cv2
import numpy as np
from utils import imshow
```

```python
img = cv2.imread('../IMG_20180908_114019.jpg')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

imshow("original", img)
imshow("gray", gray, gray=True)
```

```python
corners = cv2.goodFeaturesToTrack(gray, 4, 0.10, 10)
corners = np.int0(corners)

c_img = img.copy()

for i in corners:
    x, y = i.ravel()
    cv2.circle(c_img, (x,y), 3, 255, -1)
    
imshow("corners", c_img)
```
